```python
import transformers
import torch
```

```python
model_id = "unsloth/llama-3-8b-Instruct-bnb-4bit"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs = {
        "torch_dtype":torch.float16,
        "low_cpu_mem_usage":True,
    }
)
```

```python
def generate_output(messages):

  prompt = pipeline.tokenizer.apply_chat_template(
      messages,
      tokenize=False,
      add_generation_prompt = True
  )

  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = pipeline(
      prompt,
      max_new_tokens = 1024,
      eos_token_id = terminators,
      do_sample = True,
      temperature = 0.6,
      top_p = 0.9
  )

  return outputs[0]["generated_text"][len(prompt):]
```

Welcome to the fascinating world of Large Language Models (LLMs)! As a teacher, I'm excited to introduce you to this cutting-edge technology.

Large Language Models are artificial intelligence (AI) systems trained on vast amounts of text data to generate human-like language. They're called "Large" because they're designed to process and learn from enormous amounts of text, often exceeding the size of a small library.

Imagine a super-smart, language-loving robot that can:

1. **Understand** vast amounts of text, from books to articles, and even conversations.
2. **Generate** new text, such as stories, articles, or even entire books, that are coherent and natural-sounding.
3. **Respond** to questions, engage in conversations, or even create new content based on a given topic or prompt.

LLMs are trained using a process called **deep learning**, which is a type of machine learning that's inspired by the human brain. They're typically built using neural networks, which are composed of many layers of interconnected nodes (or "neurons").

Here's a simplified analogy to help you understand how LLMs work:

Think of an LLM as a super-powerful, language-translation machine. Imagine you're trying to translate a complex text from one language to another. A human translator would need to read the original text, understand its meaning, and then translate it into the target language. An LLM, on the other hand, can do this process automatically, using its vast knowledge of language patterns and structures.

When you input a prompt or question into an LLM, it:

1. **Reads** the input text, just like a human would.
2. **Understands** the meaning and context of the input.
3. **Generates** a response, which is a new piece of text that's relevant to the input.

LLMs have many practical applications, such as:

* **Content creation**: LLMs can generate articles, blog posts, or even entire books on a given topic.
* **Chatbots**: LLMs can power conversational AI systems, enabling them to engage in natural-sounding conversations.
* **Language translation**: LLMs can translate text from one language to another, helping to break down language barriers.
* **Research**: LLMs can assist in research tasks, such as summarizing large documents, identifying key concepts, or even generating new research questions.

As you explore the world of LLMs, you'll discover their incredible potential to transform industries, from education to healthcare, and beyond!

```python
messages = [
    {'role':"system", "content":"Your are a Python, AI and ML Expert as well as an experienced data scientist"},
    {'role':"user", "content":'Explain and write me a code to predict house prices'},
]

print(generate_output(messages))
```
